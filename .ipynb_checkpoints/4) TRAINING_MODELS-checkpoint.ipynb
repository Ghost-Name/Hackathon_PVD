```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Инициализируем модели
linreg_model = LinearRegression()

rf_reg_model = RandomForestRegressor(random_state=42)

gb_reg_model = GradientBoostingRegressor(random_state=42)
```

```python
# https://stackoverflow.com/questions/41925157/logisticregression-unknown-label-type-continuous-using-sklearn-in-python

y = df['Devotion'] #нужен параметр у которого значение 0 или 1

X = df[['Размер уставного капитала объявленный', 'Численность персонала по данным ФНС.Количество', 
    'Карточка клиента (внешний источник).Индекс платежной дисциплины Значение', 
    'Карточка клиента (внешний источник).Индекс финансового риска Значение',
    'Провозная плата (период 1)', 'Провозная плата (период 2)',
    'Объем перевозок(тн) (период 1)', 'Объем перевозок(тн) (период 2)',
    'positive_action']]
```

```python
from sklearn.model_selection import train_test_split

# Разделяем данные на обучающую (80%) и тестовую (20%) выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
```

```python
# Обучение моделей на обучающей выборке
linreg_model.fit(X_train, y_train)
rf_reg_model.fit(X_train, y_train)
gb_reg_model.fit(X_train, y_train)

#logreg_model = LogisticRegression()
#logreg_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Функция для оценки производительности модели для регрессии
def evaluate_regression_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mae, mse, r2

# Оценка производительности моделей
linreg_mae, linreg_mse, linreg_r2 = evaluate_regression_model(linreg_model, X_test, y_test)
rf_mae, rf_mse, rf_r2 = evaluate_regression_model(rf_reg_model, X_test, y_test)
gb_mae, gb_mse, gb_r2 = evaluate_regression_model(gb_reg_model, X_test, y_test)

# Вывод результатов
print("Результаты оценки производительности моделей на тестовой выборке:")
print("Линейная регрессия:")
print(f"MAE: {linreg_mae:.4f}, MSE: {linreg_mse:.4f}, R²: {linreg_r2:.4f}")
print("Случайный лес:")
print(f"MAE: {rf_mae:.4f}, MSE: {rf_mse:.4f}, R²: {rf_r2:.4f}")
print("Градиентный бустинг:")
print(f"MAE: {gb_mae:.4f}, MSE: {gb_mse:.4f}, R²: {gb_r2:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные для графика
models = ['Линейная регрессия', 'Случайный лес', 'Градиентный бустинг']
mae_values = [linreg_mae, rf_mae, gb_mae]
mse_values = [linreg_mse, rf_mse, gb_mse]
r2_values = [linreg_r2, rf_r2, gb_r2]

x = np.arange(len(models))  # Места для каждой модели
width = 0.25  # Ширина столбцов

# Создание графика
fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - width, mae_values, width, label='MAE', color='#A8D8E8')
bars2 = ax.bar(x, mse_values, width, label='MSE', color='#F3E5AB')
bars3 = ax.bar(x + width, r2_values, width, label='R²', color='#D1D1E1')

# Добавление меток и заголовков
ax.set_xlabel('Модели')
ax.set_title('Сравнение производительности регрессионных моделей')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Добавление значений над столбцами
def add_value_labels(bars):
    """Добавляет значения над столбцами графика."""
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 4), 
                ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

# Показ графика
plt.tight_layout()
plt.show()
```

```python
# Предполагаем, что у вас есть колонка 'ID' в вашем DataFrame df
# Получаем предсказания от модели
predictions = gb_model.predict(X_test)  # Или другой выбранной модели

# Создаём новый DataFrame с ID и предсказанными значениями
results = pd.DataFrame({
    'ID': df.loc[X_test.index, 'ID'],  # Предположим, что в df есть колонка ID
    'Predicted_Devotion': predictions
})

# Выводим результаты
print(results[5:])
```

```python
import pandas as pd

# Загрузка файла с актуальными данными
true_values_df = pd.read_excel(r'Sample_20_people\train20_data.xlsx')

# Объединяем предсказания с реальными значениями по ID
results_with_true = results.merge(true_values_df[['ID', 'Devotion']], on='ID')

# Рассчитываем точность предсказаний
results_with_true['Error_GB'] = results_with_true['Predicted_Devotion_GB'] - results_with_true['Devotion']
results_with_true['Error_RF'] = results_with_true['Predicted_Devotion_RF'] - results_with_true['Devotion']
results_with_true['Error_LinReg'] = results_with_true['Predicted_Devotion_LinReg'] - results_with_true['Devotion']

# Можно использовать различные метрики для оценки точности, например, среднюю абсолютную ошибку (MAE)
mae_gb = (results_with_true['Error_GB'].abs()).mean()
mae_rf = (results_with_true['Error_RF'].abs()).mean()
mae_linreg = (results_with_true['Error_LinReg'].abs()).mean()

# Печатаем результаты
print("Mean Absolute Error (MAE):")
print(f"Gradient Boosting: {mae_gb}")
print(f"Random Forest: {mae_rf}")
print(f"Linear Regression: {mae_linreg}")
```

```python
Преимущества MAE:

    Интуитивность: MAE легко интерпретируется и показывает среднюю ошибку в тех же единицах, что и данные.
    Чувствительность к выбросам: MAE менее чувствительна к выбросам по сравнению с другими метриками, такими 
    как MSE (Mean Squared Error), поскольку не возводит ошибки в квадрат.
```
