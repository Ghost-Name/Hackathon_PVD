## Объёмы перевозок.xls (Уникальных ID-956):

```python
import pandas as pd
```

```python
input_file = r'Interests_Calls_Volumes\Объёмы перевозок.xls'
output_file = r'Interests_Calls_Volumes_FILTERED\Объёмы_перевозок_filtered.xlsx'

data = pd.read_excel(input_file, header=1)

# Убираем лишние пробелы в названиях колонок и переименовываем
data.columns = data.columns.str.strip()
data.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)

# Определяем текущую дату
current_year = 2024
current_month = 10  

# Последние 12 доступных месяцев
available_months = []

# Проходим по месяцам начиная с текущего и идем назад во времени
for year in range(current_year, 2019, -1):  # Предположим, что данные начинаются с 2020 года
    for month in range(12, 0, -1):
        month_str = f'{year}/{month:02d}'
        if month_str in data.columns and f'{month_str}.1' in data.columns:
            available_months.append(month_str)
            if len(available_months) == 12:
                break
    if len(available_months) == 12:
        break

if len(available_months) < 12:
    print("Недостаточно месяцев для анализа. Найденные месяцы:", available_months)
else:
    results = []

    # Определяем периоды
    period_1 = available_months[:6]  # Первые 6 месяцев
    period_2 = available_months[6:12] # Последние 6 месяцев

    # Проходим по всем строкам данных
    for _, row in data.iterrows():
        id_value = row['ID']
        
        # Собираем данные для первых 6 месяцев
        провозная_плата_1 = row[period_1].sum() if all(pd.notna(row[month]) for month in period_1) else None
        объем_перевозок_1 = row[[f'{month}.1' for month in period_1]].sum() if провозная_плата_1 is not None else None

        # Собираем данные для последних 6 месяцев
        провозная_плата_2 = row[period_2].sum() if all(pd.notna(row[month]) for month in period_2) else None
        объем_перевозок_2 = row[[f'{month}.1' for month in period_2]].sum() if провозная_плата_2 is not None else None

        results.append({
            'ID': id_value,
            'Провозная плата (период 1)': провозная_плата_1,
            'Объем перевозок(тн) (период 1)': объем_перевозок_1,
            'Провозная плата (период 2)': провозная_плата_2,
            'Объем перевозок(тн) (период 2)': объем_перевозок_2
        })

    # Превращаем результаты в DataFrame
    result_df = pd.DataFrame(results)

    # Убираем строки с отсутствующими данными
    result_df.dropna(subset=['Провозная плата (период 1)', 'Объем перевозок(тн) (период 1)', 
                             'Провозная плата (период 2)', 'Объем перевозок(тн) (период 2)'], inplace=True)

    result_df.to_excel(output_file, index=False)
    print("Фильтрация завершена. Результат сохранен в:", output_file)

print("Используются месяцы: ", available_months)
```

```python
# убирает непотребства 
file_path = r'Interests_Calls_Volumes_FILTERED\Объёмы_перевозок_filtered.xlsx'
df = pd.read_excel(file_path, header=1)  # header=1 пропускает первую строку

df.columns = [
    'ID', 
    'Провозная плата (период 2)', 
    'Объем перевозок(тн) (период 2)', 
    'Провозная плата (период 1)', 
    'Объем перевозок(тн) (период 1)'
]

# Суммирование по уникальному ID
df_summary = df.groupby('ID').sum().reset_index()

# Сохранение результата в новый файл, если нужно
output_path = r'Interests_Calls_Volumes_FILTERED\Объёмы_перевозок_periods.xlsx'
df_summary.to_excel(output_path, index=False)
```

## создание дополнительного параметра positiv_action

```python
activity_file_path = 'Interests_Calls_Volumes/Обращения.xls'
activity_df = pd.read_excel(activity_file_path)


filtered_data_path = 'Sample_20_people/filtered_data.xlsx'
filtered_df = pd.read_excel(filtered_data_path)

topics_of_interest = [
    "Цифровые сервисы",
    "Электронный обмен документами",
    "Перевозка грузов и порожних вагонов",
    "Оформление документов связанных с перевозкой грузов",
    "Заключение договоров",
    "Финансовые расчеты",
    "Перевозка грузов в контейнерах",
    "Благодарности",
    "Перевозка грузов в рефрижераторных секциях ИВ-термосах вагонах-термосах",
    "Перевозка сборной/мелкой партии грузов / разовая"
]


filtered_activity_df = activity_df[activity_df['Тема вопроса'].isin(topics_of_interest)]
positive_action_counts = filtered_activity_df['ID'].value_counts()

filtered_df['positive_action'] = filtered_df['ID'].map(positive_action_counts).fillna(0)


output_path = 'Sample_20_people/filtered_data_with_positive_action.xlsx'
filtered_df.to_excel(output_path, index=False)
```
